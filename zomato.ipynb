```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("Zomato data .csv")
```

```python
df
```

```python
df.info()
```

```python
def conv(value):
    value = str(value).split("/")
    value = value[0]
    return float(value)
    
df["rate"]= df["rate"].apply(conv)
df
```

```python
df.head(20)
```

```python
df.isnull().sum()
```

```python
df.describe()
```

```python
df.duplicated()
```

```python

sns.countplot(x= "listed_in(type)", data = df)
plt.title("majority of customer order from")
plt.show()
```

Most of the customer order from dining restaurant 

```python
sns.barplot(x = "listed_in(type)", y = "votes",data = df, hue="online_order")
plt.title("vote recived by each type of restaurant")
plt.show()
```

Maximum votes is recieved by others types of restuarent 

```python
plt.figure(figsize = (9,4))
sns.histplot(x = "rate", data = df, color = "darkorange")
plt.title("rating recived by restaurant")
plt.grid()
plt.show()
```

majority of restuarent recived rating of 3.5  

```python
plt.figure(figsize = (8,3))
sns.countplot(x = "approx_cost(for two people)", data = df, hue ="online_order" )
plt.title("average spend of couple")
plt.grid()
plt.show()
```

Most of the couple spend rs.300 offline and around 500-600 online 

```python
plt.figure(figsize=(6,6))
sns.boxplot(x ="online_order",y = "rate",data = df)
plt.title("online or ofline rating ")
plt.show()
```

Online order recived a maximum number of rating 

```python
sns.countplot(x = "listed_in(type)", data = df, hue = "online_order")
plt.title("maximum ofline order reciverd by")
plt.grid()
plt.show()
```

maximum ofline order is recived by Dining type of restaurant
